Generating solar power relies on certain conditions like weather, the number of sensors, and how the data is gathered through an inverter. The data is sorted into two files with seven columns each, one containing the data for the power that is generated, and the other for the weather.
From the power generated dataset we have:

| **DATE_TIME**                      | **PLANT_ID**          | **SOURCE_KEY**                                                                                    | DC_POWER                           | AC_POWER                                | DAILY_YIELD                         | TOTAL_YIELD                                    |
|------------------------------------|-----------------------|---------------------------------------------------------------------------------------------------|------------------------------------|-----------------------------------------|-------------------------------------|------------------------------------------------|

| **DATE_TIME**                      | **PLANT_ID**          | **SOURCE_ID**                   | AMBIENT_TEMPERATURE                              | MODULE_TEMPERATURE                                                                                      | IRRIDATION             |
